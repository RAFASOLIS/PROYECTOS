
<h1><u> Conversor de archivos JSON</u></h1><br/>
El Notebook abre una ventana en la que aparecen 2 botones. El botón "Importar" abre un explorador de archivos que te permite realizar la búsqueda del fichero navegando por el sistema de carpetas del PC del usuario y seleccionar el fichero con extensión .json. El botón exportar transforma el fichero seleccionado en un fichero con la misma información pero en formato Excel, incluyendo una columna adicional con el nombre del fichero de origen.
<h2>Instrucciones</h2>
    <ol>
    <li>Seleccionamos la siguiente casilla In[] del notebook y pulsamos "Run".</li>
    <li>En la ventana que se abre pulsamos el botón "Importar" y seleccinamos el fihhero json.</li>
    <li>Pulsamos el botón exportar y nos generará un fichero en la carpera "Descargas" con el mismo nombre que el fichero seleccionado pero con formato excel.</li>
    </ol>

```python
#CASILLA In[] A SELECCIONAR PARA LA EJECUCIÓN

#Importamos librerías

from tkinter import *
from tkinter.filedialog import askopenfilename
from getpass import getuser
import pandas as pd
import os

#Definimos variables globales
path="C:/Users/"+getuser()+"/Downloads/"
datos=""
message=""
filename=""

#Definimos funciones
   
def exportar():
    global message

    try:
        datos["origen"]=filename.name.split(path)[1]
        filename.close()
        datos.to_excel(path+"/"+filename.name.split(path)[1].split(".json")[0]+".xlsx")
        message="Datos exportados"
    except:
        message="Error de exporcación"
        
    
def importar():
    global datos
    global message
    global filename
    try:
        file_up=askopenfilename(initialdir=path,title="examinar",filetypes = (("JSON files", "*.json"),("All files", "*.*"))) 
        filename=open(file_up)
        
        datos=pd.read_json(file_up,encoding="utf-8")
        message="Datos importados"
    except:
        message="Error de importación"

#Definimos interfaz gráfica        
ventana=Tk()
ventana.title("Conversor de fichero JSON")
ventana.geometry("400x200")


l_descript=Label(ventana,text="\n1. Selecciona el fichero JSON. ubicado en 'Descargas'.\n 2.Exporta el fichero a Excel en la misma ruta.")



bt_importar=Button(ventana,text="Importar JSON",command=importar)
bt_exportar=Button(ventana,text="Exportar EXCEL",command=exportar)
l_msg=Label(ventana,text=message)

l_copy=Label(ventana,text="\n\n© Rafael Solís López")

l_descript.pack()
bt_importar.pack()
bt_exportar.pack()
l_msg.pack()
l_copy.pack()

mainloop()
```
